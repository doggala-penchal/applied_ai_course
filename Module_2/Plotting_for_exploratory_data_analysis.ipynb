```python
# One dimentional vector is called Scalar.[0] or [1], etc
# all vectors other than scalars are called Vectors. [1, 2, 3, 4, 5]
# Every observation in data set is called observation or certain dimension vector. [0,1,2,3,4,5]
# First gain domain knowledge so that you can get clear intuition about the problem.
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sea 
```

```python
x = np.array([1,2,3,4,5])
y = np.array([6,7,8,9,10])
plt.plot(x, y, 'ro')
plt.show()
```

```python
data = pd.read_csv('iris.csv')
data
```

```python
data.shape
```

```python
data.columns
```

```python
data['species'].value_counts() # .value_counts() gives different values in that column
```

```python
plt.scatter(data['sepal_length'], data['sepal_width'])
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')
plt.title('sepal_length vs sepal_width')
```

```python
sea.set_style('whitegrid')
sea.FacetGrid(data, hue= 'species', size = 5).map(plt.scatter, 'sepal_length', 'sepal_width').add_legend()

plt.show()
```

```python
plt.close
sea.set_style('whitegrid')
sea.pairplot(data, hue= 'species', size= 4)
plt.show()
```
