```python
import matplotlib.pyplot as plt
```

These are some commmand line functions which make matplotlib work as matlab.

```python
plt.plot([2, 4, 6, 8, 5, 2, 7])
plt.xlabel("Indices")
plt.ylabel("Numbers")
plt.title("Plot of Numbers vs Indices")
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.xlabel('X_values')
plt.ylabel('Y_values')
plt.grid() # grid helps in reading numbers easily
plt.show()
```

```python
plt.plot([1,2,3,4,5,6,7])
plt.title("Plot of the x-axis and y-axis")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show() # if no y values are gives then they will take the indcess as y values.
```

```python
import numpy as np 
T = np.arange(0., 5., 0.2)
T
```

```python
plt.plot(T, T**2, 'b--', label = '^2')
plt.plot(T, T**2.1, 'gs', label = '^2.1')
plt.plot(T, T**2.5, 'r^', label = '^2.5')
plt.legend() # this is used to represent labels in the plot.and legend explains to understand which plot belongs to what.
plt.grid()
plt.show()
```

```python
x = [1,2,3,4,5]
y = [3,1,7,9,5]
plt.plot(x, y, linewidth = 15.0)
plt.show()
```

```python
x1 = [1,2,3,4,5]
y1 = [2,4,6,8,10]
x2 = [1,3,5,7,9]
y2 = [9,7,5,3,1]
lines = plt.plot(x1, y1, x2, y2)
plt.setp(lines[0], color = 'r', linewidth = 5.0)
plt.setp(lines[1], color = 'b', linewidth = 10.0)
plt.grid()
plt.xlabel("x-axis")
plt.xlabel('y-axis')
plt.title('Simple_example')
```

Working with multiple plots and axis

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)
t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)
plt.figure(1)
plt.subplot(211)
plt.grid()
plt.plot(t1, f(t1), 'b-')
plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

```python
plt.figure(1)
plt.figure(211)
plt.plot([1,4,9,16,25])
plt.figure(212)
plt.plot([25,16,9,4,1])
plt.show()
plt.figure(2)
plt.plot([1,13,15,17,19])
plt.show()
```
